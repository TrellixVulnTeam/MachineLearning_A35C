```python
#將數據分成訓練跟測試
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.4, random_state =123) #random_state為隨機種子
```

```python
#pearson : 皮爾森相關係數
#約=1 完全相關 (Perfect correlated)
#0.7~0.99 高度相關 (Highly correlated)
#0.4~0.69 中度相關 (Moderately correlated)
#0.1~0.39 低度相關 (Modestly correlated)
#0.01~0.09 接近無相關 (Weakly correlated)
#約=0 無相關
#kendall：用於反映分類變數相關性的指標，即針對無序序列的相關係數，非正太分佈的資料
#spearman：非線性的，非正太分析的資料的相關係數
cor = used_1.corr(method='pearson')
#Correlation with output variable
cor_price = abs(cor["price"])
#Selecting highly correlated features
relevant_features = cor_price[cor_price>0.1]
relevant_features.sort_values(ascending=False)
```

```python
'''從皮爾森相關係數得知有中度相關影響二手車車價的條件分別是年份(year)和電動座椅(power chair)，而有低度相關影響車價的分別有
定速(cruise control)和免鑰匙啟動(push start)。'''
#另外車子的品牌(brand)和車子的所在地(city)也會影響車價
used_3 = used_1[['year', 'power_chair', 'cruise_control', 'push_start','brand','city','price']]
used_3
```
