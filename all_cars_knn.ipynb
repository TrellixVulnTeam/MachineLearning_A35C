```python
# 將 price 欄標準化
from sklearn import preprocessing
# 先取出 price欄
price = pd.DataFrame(df['price'])
price_scaled = preprocessing.scale(price)
price_scaled
```

```python
#過濾outliner => 只拿三個標準差之內的資料
df_price_scaled = pd.DataFrame(price_scaled)

# 使用 concat 合併時，他預設的 join 模式是 'outer'，會直接把沒有的資料用 NaN 代替
# axis=1 =>columns合併  => R:cbind()
# axis=0 =>rows合併 => R:rbind()

df = pd.concat([df, df_price_scaled],axis=1,join='inner')
# 只取標準差在3之內的數據

df = df[df[abs(0)]< 3]
df
```

```python
#篩選掉欄位
df.drop(columns=[0, 'price'], axis=1)
df
```

```python
#pearson : 皮爾森相關係數
#約=1 完全相關 (Perfect correlated)
#0.7~0.99 高度相關 (Highly correlated)
#0.4~0.69 中度相關 (Moderately correlated)
#0.1~0.39 低度相關 (Modestly correlated)
#0.01~0.09 接近無相關 (Weakly correlated)
#約=0 無相關
#kendall：用于反映分类变量相关性的指标，即针对无序序列的相关系数，非正太分布的数据
#spearman：非线性的，非正太分析的数据的相关系数
cor = df.corr(method='pearson')
#cor
# Correlation with output variable
cor_price = abs(cor["price"])
#Selecting highly correlated features
relevant_features = cor_price[cor_price>0.1]
relevant_features.sort_values(ascending=False)
```

```python
#重要特徵x變數 (皮爾斯相關係數)
# year                    0.535858
# power_chair             0.404331
# cruise_control          0.392337
# push_start              0.321097
# hid                     0.285197
# reversing_video         0.253271
# climate_control         0.247691
# satellite_navigation    0.220039
# dvd                     0.180497
# vcd                     0.145303

#放入除了car_id、model、url、pic_url之外的欄位
df = df[['brand','year','engine_displacement','op_type','price','used_new','km','cruise_control','hid','cd','vcd','dvd','reversing_rader','reversing_video','satellite_navigation','climate_control','push_start','anti_theft_device','power_chair','power_window','city']]
df
```

```python
from decimal import *

#將類別型欄位轉One-Hot Encoding
#轉完One-Hot Encoding後將資料 => df_dummy
df_dummy = pd.get_dummies(df)
df_dummy
#隨機抽樣
#DataFrame.sample(n=None, frac=None, replace=False, weights=None, random_state=None, axis=None)
#訓練資料集占全部資料量的比率
train_ratio = 0.7


#train dataset的抽樣數量
count_train = int(Decimal(cnt_total)*Decimal(train_ratio))
# print(count_train)
count_test = cnt_total - count_train
# print(count_test)

#抽樣
#replace：是否为有放回抽样，取replace=True时为有放回抽样 預設=False
#random_state類似R語言的seed，固定每次抽樣，避免每次跑的結果不一

#Training Set抽樣
#df_train = df_dummy.sample(n=count_train,replace=True,random_state=1)
df_train = df_dummy.sample(n=count_train,replace=True)
df_train

#Testing Set抽樣
#df_test = df_dummy.sample(n=count_test,replace=True,random_state=1)
df_test = df_dummy.sample(n=count_test,replace=True)
df_test

#將Dataframe轉換成Series
#y_train => 取出price series
y_train = df_train["price"].values
#y_train

#y_test => 取出price series
y_test = df_test["price"].values
#y_test

#exclude column:price
df_train_exclude_price = df_train.drop(["price"], axis=1)
x_train = df_train_exclude_price.values
#x_train

df_test_exclude_price = df_test.drop(["price"], axis=1)
x_test = df_test_exclude_price.values
#x_test
```

```python
#將資料集做標準化作業
#https://blog.csdn.net/cymy001/article/details/79112818
from sklearn.preprocessing import StandardScaler
ss_X=StandardScaler()
ss_y=StandardScaler()

x_train=ss_X.fit_transform(x_train)
x_test=ss_X.transform(x_test)
#x_train

y_train=ss_y.fit_transform(y_train.reshape(-1, 1))
y_test=ss_y.transform(y_test.reshape(-1, 1))
```

```python
from sklearn import neighbors
import matplotlib.pyplot as plt

#分成5群
n_neighbors = 5

#https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html
#‘uniform’ : uniform weights. All points in each neighborhood are weighted equally.
weights = 'uniform'    
knn_uniform = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
y_uniform_pred = knn_uniform.fit(x_train, y_train).predict(x_test)

#‘distance’ : weight points by the inverse of their distance. 
#in this case, closer neighbors of a query point will have a greater influence than neighbors which are further away.    
weights = 'distance'    
knn_distance = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
y_distance_pred = knn_distance.fit(x_train, y_train).predict(x_test)
```

```python
#計算ACC
from sklearn.metrics import r2_score
#weights= 'uniform'
print("KNN weights='uniform' ACC=",r2_score(y_test, y_uniform_pred,multioutput='variance_weighted'))
#weights= 'distance'
print("KNN weights='distance' ACC=",r2_score(y_test, y_distance_pred,multioutput='variance_weighted'))
```
