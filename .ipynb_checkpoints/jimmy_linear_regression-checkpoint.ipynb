```python
import pandas as pd

import mysql.connector

conn = mysql.connector.connect(db="iiicars",
                               host="139.162.104.154",
                               user="jimmy",
                               passwd="Chiu!@#$%lin1024"
                               )
#除了cars_id、model、used_new、address、source、url和pic_url沒有抓
sql = "SELECT brand,year,engine_displacement,op_type,price,km,cruise_control,hid,cd,vcd,dvd,reversing_rader,reversing_video,satellite_navigation,climate_control,push_start,anti_theft_device,power_chair,power_window,city FROM cars"

#設定最多顯示的column數量
pd.set_option('display.max_columns', 500)

df = pd.read_sql_query(sql,conn) 
df.head()
```

```python
# 顯示空值列
df[df.isnull().values==True]
```

```python
# 將 price 欄標準化
from sklearn import preprocessing
# 先取出 price 欄
price = pd.DataFrame(df['price'])
price_scaled = preprocessing.scale(price)
price_scaled
```

```python
# 將 price_scaled欄位concat data frame，並且只保留標準差<3的值
price_scaled = pd.DataFrame(price_scaled)
df = pd.concat([df, price_scaled],axis=1,join='inner')
df = df[df[abs(0)]<3]
df.drop(columns=[0],axis=1,inplace=True)
df
```

```python
df_encoded = pd.get_dummies(df)
df_encoded
```

```python
x = df_encoded.drop("price", axis=1)
y = df_encoded['price']
```

```python
#將數據分成訓練跟測試
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.3, random_state =123) #random_state為隨機種子
#x_train type=> DataFrame
print(type(x_train))
```

```python
#導出one-hot encoding的encoder
#給測試資料集使用
from sklearn.preprocessing import OneHotEncoder
import joblib
import numpy as np

x_train_np = np.array(x_train)

enc = OneHotEncoder(categories = "auto", handle_unknown = "ignore")
df_hotEncoded = enc.fit_transform(x_train_np)
print(df_hotEncoded)
joblib.dump(enc, "linear_encoder.pkl")
```

```python
#初始化線性回歸模型，並對線性回歸模型進行訓練
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#儲存model檔案
import joblib

joblib.dump(regressor, open("linear_model.pickle", "wb"))
```

```python
#預測測試結果
y_pred = regressor.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
